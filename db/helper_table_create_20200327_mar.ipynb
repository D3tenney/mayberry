```python
import pandas as pd
# county
first = True

pull_date = '20200314'
vf_dir = '/home/dtenney/Documents/gitprojects/mayberry/voterfile/'
output_dir = './helper_tables/helper_tables_'+pull_date+'/'

for chunk in pd.read_csv(vf_dir + 'voterfile_nc_'+pull_date+'.csv', chunksize=100000):
    
    county_df = chunk[['county_id', 'county_desc']].drop_duplicates().reset_index(drop=True)
    ethnic_df = chunk[['ethnic_code']].drop_duplicates().reset_index(drop=True)
    race_df = chunk[['race_code']].drop_duplicates().reset_index(drop=True)
    party_df = chunk[['party_cd']].drop_duplicates().reset_index(drop=True)
    precinct_df = chunk[['county_id', 'precinct_abbrv', 'precinct_desc']].drop_duplicates().reset_index(drop=True)
    status_df = chunk[['status_cd', 'voter_status_desc']].drop_duplicates().reset_index(drop=True)
    reason_df = chunk[['reason_cd', 'voter_status_reason_desc']].drop_duplicates().reset_index(drop=True)

    if first is True:
        county = county_df
        ethnic = ethnic_df
        race = race_df
        party = party_df
        precinct = precinct_df
        status = status_df
        reason = reason_df

        first = False
    else:
        county = county.append(county_df).reset_index(drop=True)
        ethnic = ethnic.append(ethnic_df).reset_index(drop=True)
        race = race.append(race_df).reset_index(drop=True)
        party = party.append(party_df).reset_index(drop=True)
        precinct = precinct.append(precinct_df).reset_index(drop=True)
        status = status.append(status_df).reset_index(drop=True)
        reason = reason.append(reason_df).reset_index(drop=True)
        
county = (county.rename(columns={'county_id': 'id', 'county_desc': 'name'})
          .drop_duplicates()
          .reset_index(drop=True)
          .sort_values('id'))
ethnic = (ethnic.rename(columns={'ethnic_code': 'code'})
          .drop_duplicates().reset_index(drop=True).sort_values('code'))
race = (race.rename(columns={'race_code': 'code'})
        .drop_duplicates().reset_index(drop=True).sort_values('code'))
party = (party.rename(columns={'party_cd': 'cd'})
         .drop_duplicates().reset_index(drop=True).sort_values('cd'))
precinct = (precinct.rename(columns={'precinct_abbrv': 'abbrv', 'precinct_desc': 'description'})
            .drop_duplicates(subset=['county_id', 'abbrv'])
            .reset_index(drop=True).sort_values(['county_id', 'abbrv']))
status = (status.rename(columns={'status_cd': 'cd', 'voter_status_desc': 'description'})
          .drop_duplicates().reset_index(drop=True).sort_values('cd'))
reason = (reason.rename(columns={'reason_cd': 'cd', 'voter_status_reason_desc': 'description'})
          .drop_duplicates().reset_index(drop=True).sort_values('cd'))
```
