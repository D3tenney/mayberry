```python
import os
from sqlalchemy import create_engine
from dotenv import load_dotenv
import pandas as pd

load_dotenv(override=True)

# DATABASE CONNECTION
PG_USER = os.environ.get("PG_USER")
PG_PASS = os.environ.get("PG_PASS")
PG_HOST = os.environ.get("PG_HOST")
PG_DB = os.environ.get("PG_DB")
PG_PORT = os.environ.get("PG_PORT")

conn_string = "postgresql://{0}:{1}@{2}:{3}/{4}".format(
    PG_USER, PG_PASS, PG_HOST, PG_PORT, PG_DB)
engine = create_engine(conn_string)
conn = engine.connect()
```

```python
first_run = True
for chunk in pd.read_csv('./voterfile_nc_20200201.csv', chunksize=10000):
    if first_run:
        print('first run start')
    df = chunk[['ncid',
                'voter_reg_num',
                'first_name',
                'middle_name',
                'last_name',
                'status_cd',
                'reason_cd',
                'absent_ind',
                'res_street_address',
                'res_city_desc',
                'state_cd',
                'zip_code',
                'mail_addr1',
                'mail_addr2',
                'mail_addr3',
                'mail_addr4',
                'mail_city',
                'mail_state',
                'mail_zipcode',
                'full_phone_number',
                'drivers_lic',
                'race_code',
                'ethnic_code',
                'party_cd',
                'gender_code',
                'birth_year',
                'birth_age',
                'birth_state',
                'registr_dt',
                'county_id',
                'precinct_abbrv',
                'cong_dist_abbrv',
                'nc_senate_abbrv',
                'nc_house_abbrv']]

    df.to_sql('voter_staging',
              if_exists='append',
              con=conn,
              method='multi',
              index=False,
              chunksize=10000)
    if first_run:
        print('first run inserted')
        first_run = False
```

```python
conn.close()
engine.dispose()
```
