```python
import pandas as pd
```

```python
status = []

for chunk in pd.read_csv('./voterfile_nc_20200201.csv', chunksize=100000):
    unique = chunk['status_cd'].unique()
    
    for item in unique:
        if item not in status:
            status.append(item)
            
status
```

```python
first = True

for chunk in pd.read_csv('./voterfile_nc_20200201.csv', chunksize=100000):
    unique = chunk[['reason_cd', 'voter_status_reason_desc']].drop_duplicates().reset_index(drop=True)
    
    if first is True:
        df = unique
        first = False
    else:
        df = df.append(unique).reset_index(drop=True)
    
            
df = df.drop_duplicates().reset_index(drop=True)
df
```

```python
df.rename(columns={'reason_cd': 'cd', 'voter_status_reason_desc': 'description'}, inplace=True)
df.to_csv('./helper_tables_20200217/status_reason.csv', index=False)
```
